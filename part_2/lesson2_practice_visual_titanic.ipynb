**Считываем обучающую выборку.**

```python
train_df = pd.read_csv("C:\\Users\\Zloy\\mlcours\\titanic_train.csv", 
                       index_col='PassengerId') 
```

**Выкинем признак `Cabin`, а потом – все строки, где есть пропуски.**

```python
train_df = train_df.drop('Cabin', axis=1).dropna()
```

**Постройте попарные зависимости признаков `Age`, `Fare`, `Pclass`, `Sex`, `SibSp`, `Parch`, `Embarked` и `Survived`. (метод `scatter_matrix ` Pandas или `pairplot` Seaborn).**

```python
features = ['Age', 'Fare', 'Pclass', 'Sex', 'SibSp', 'Parch', 'Embarked' , 'Survived']
sns.pairplot(train_df[features])
```

**Как плата за билет (`Fare`) зависит от класса каюты (`Pclass`)? Постройте boxplot.**

```python
p1_std, p1_mn = train_df[train_df['Pclass']==1]['Fare'].std(), train_df[train_df['Pclass']==1]['Fare'].mean()
p2_std, p2_mn = train_df[train_df['Pclass']==2]['Fare'].std(), train_df[train_df['Pclass']==2]['Fare'].mean()
p3_std, p3_mn = train_df[train_df['Pclass']==3]['Fare'].std(),train_df[train_df['Pclass']==3]['Fare'].mean()
train_df[train_df['Pclass']==3].describe(), p3_std
```

**Такой boxplot получается не очень красивым из-за выбросов.** 

**Опционально: создайте признак `Fare_no_out` (стоимости без выбросов), в котором исключаются стоимости, отличающиеся от средней по классу более чем на 2 стандартных отклонения. Важно: надо исключать выбросы именно в зависимости от класса каюты. Иначе исключаться будут только самые большие (1 класс) и малые (3 класс) стоимости.**

```python
fare_pclass1_no_out = train_df[(train_df['Pclass']==1) & ( abs(train_df['Fare'] - p1_mn) < 2*p1_std)]
fare_pclass2_no_out = train_df[(train_df['Pclass']==2) & ( abs(train_df['Fare'] - p2_mn) < 2*p2_std)]
fare_pclass3_no_out = train_df[(train_df['Pclass']==3) & ( abs(train_df['Fare'] - p3_mn) < 2*p3_std)]
train_df_no_outs = fare_pclass1_no_out.append(fare_pclass2_no_out).append(fare_pclass3_no_out)
sns.boxplot(x='Pclass', y='Fare', data=train_df_no_outs)
```

**Каково соотношение погибших и выживших в зависимости от пола? Отобразите c помощью Seaborn.countplot c аргументом hue.**

```python
sns.countplot(x='Sex', hue='Survived', data=train_df)
```

**Каково соотношение погибших и выживших в зависимости от класса каюты? Отобразите c помощью Seaborn.countplot c аргументом hue.**

```python
sns.countplot(x='Pclass', hue='Survived', data=train_df)
plt.legend(title='Survived', loc='upper left', labels=['No', 'Yes']);
```

**Как факт выживания зависит от возраста пассажира? Проверьте (графически) предположение, что молодые чаще выживали. Пусть, условно, молодые - младше 30 лет, пожилые – старше 60 лет.**

```python
sub_df = train_df[(train_df['Age'] < 30) | (train_df['Age'] > 60)]
train_df['Young'] = (sub_df["Age"] < 60).astype("int64")
sns.countplot(x='Young', hue='Survived', data=train_df[(train_df['Young']==1) | (train_df['Young']==0)])
plt.legend(title='Survived',loc='upper left', labels=['No','Yes']);
```
